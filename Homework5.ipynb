```python
import pandas as pd;
data = pd.read_csv("source_data/datasets_38396_60978_charcters_stats.csv").drop_duplicates();
data
```

```python
from sklearn.manifold import TSNE

data2 = data[['Intelligence', 'Strength', 'Speed', 'Durability', 'Power', 'Combat']]

projection = TSNE(n_components=2).fit_transform(data2);
lowd = pd.DataFrame(projection, columns=["X1","X2"]); 
lowd['Name']=data['Name']
lowd
```

```python
tsne = pd.merge(data, lowd, how='outer', on='Name')

tsne.to_csv("derived_data/tsne_characters.csv")
tsne
```

```python
from plotnine import *; 
p = (ggplot(tsne, aes("X1", "X2")) +
  geom_point(aes(color="Alignment")))
```

```python
ggsave(p, 'figures/tsne.png')
```
